GRAPH SALES BY YEAR

```python
# fix date data type
df['Ship Date'] = df['Ship Date'].str.strip()
df['Ship Date'] = pd.to_datetime(df['Ship Date'],dayfirst=True, errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df['Ship Date'] = df['Ship Date'].dt.year
```

```python
# Graph
count = df['Ship Date'].value_counts()
sns.barplot(x=count.index, y=count.values, palette='viridis')
plt.title('Ship date')
plt.xlabel('Year')
plt.ylabel('Frequent')
plt.show()
```

7. PRODUCT SALES BY YEAR

```python
# Graph
year = df['Ship Date'].unique()
grafic = []
for i in year:
    year = df[df['Ship Date'] == i]
    sales_sum = year['Sales'].sum()
    grafic.append ({'x' : i, 'y' : sales_sum})
    
grafic = pd.DataFrame(grafic)
sns.barplot(x='x', y='y',data=grafic, palette='viridis')
plt.title('Quantity Graph')
plt.xlabel('Year')
plt.ylabel('Sum')
plt.show()
```

* Beradasarkan grafik diatas bahwa terjadi penurunan penjualan (tren down) yang signifikan pada tahun 2019

8. PURCHASING BY CUSTOMER STATES 

```python
c = df['State'].unique()
for i in c:
    count = df[df['State']== i]
    counts = count['Ship Date'].value_counts()
    sns.barplot(x=counts.index, y=counts.values, palette='viridis')
    plt.title(f'Sales in {i} by Year')
    plt.xlabel(f'{i} State')
    plt.ylabel('Frequent')
    plt.show()
```

* Berdasarkan grafik diatas dapat dilihat bahwa semua negara pelanggan memiliki tren pembelian yang menurun

9. TOP 5 PRODUCT SALES BY YEAR

```python

count = df['Product Name'].value_counts().nlargest(5)
count = (count.index).unique()
for i in count:
    counts = df[df['Product Name'] == i]
    cons = counts['Ship Date'].value_counts()
    sns.barplot(x=cons.index, y=cons.values, palette='viridis')
    plt.title(f'Grafic Product {i}')
    plt.xlabel('Year')
    plt.ylabel('Counts')
    plt.show()
```

* Berdasarkan grafik diatas bahwa 5 produk dengan tren tertinggi mengalami penurununan jumlah penjualan bahan sebagian tidak terjual
